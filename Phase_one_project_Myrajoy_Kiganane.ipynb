2. Total Injuries vs. Number of Engines

lets come up with a Scatter plot that examines the relationship between the number of engines and total injuries.

This will help us establish if the number of engines  correlate with the severity of injuries in an incident and if single-engine aircraft are involved in more severe incidents with comparison to multi-engine aircraft

```python
# Calculate Pearson correlation
correlation = df['Total.Minor.Injuries'].corr(df['Number.of.Engines'])

# Print the result
print(f"Pearson Correlation Coefficient: {correlation:.2f}")
```

```python
# import seaborn libraries 
import seaborn as sns
#plotting the scatter plot 
sns.scatterplot(x=df["Number.of.Engines"], y=df["Total.Minor.Injuries"])
#inserting the title of the scatter plot
plt.title("Number of Engines vs. Injuries")
#labeling the x axis 
plt.xlabel("Number of Engines")
#labeling the y axis 
plt.ylabel("Total Minor Injuries")

plt.show()
```

3. Aviation Incidents Over Time

lets come up with a line chart that examines Aviation incidents over time 
This will help us establish if aviation incidents increasing or decreasing over time, if there specific years with a spike or drop in incidents and we identify long-term safety trends in aviation

```python
# Convert Event.Date to datetime format
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year
```

```python
# Plot trend of incidents over time
plt.figure(figsize=(12,6))
df.groupby('Year').size().plot(kind='line', marker='o', color='red')
#labelling the title 
plt.title("Trend of Aviation Incidents Over the Years")
#labeling the y axis 
plt.ylabel("Number of Incidents")
#labeling the x axis 
plt.xlabel("Year")

plt.show()
```

4. Distribution of Incidents by Number of Engines
lets come up with a histogram that examines the Distribution of Incidents by Number of Engines

This will help us establish the type of aircraft is involved in incidents most often and help us identify whether incidents are more common in aircraft with fewer or more engines.


```python
#plotting the histogram 
plt.figure(figsize=(10,6))
plt.hist(df['Number.of.Engines'], bins=5, color='orange', edgecolor='black')
#labeling the title
plt.title("Distribution of Incidents by Number of Engines")
#labeling the x axis 
plt.xlabel("Number of Engines")
#labeling the y axis 
plt.ylabel("Number of Incidents")
#inserting the mean line in the histogram
plt.axvline(df['Number.of.Engines'].mean(), color='black', linestyle='dashed', linewidth=2)

plt.show()
```
